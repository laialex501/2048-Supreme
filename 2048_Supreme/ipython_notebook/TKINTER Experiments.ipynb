```python
root = tk.Tk()

def printName():
    print("Hello there")
    
def printMeal(event):
    print("Another greeting")

# technique 1 for binding button to function
button1 = tk.Button(root, text="Click me for a greeting", command=printName)
button1.grid(row=0)
# technique 2 for binding button to function
button2 = tk.Button(root, text="Another button")
# bind requires function take in an event
button2.bind("<Button 1>", printMeal)
button2.grid(row=1)

root.mainloop()
```

```python
root = tk.Tk()

# Not limited to binding 1 function to 1 widget
# Bind multiple events to a widget

def leftClick(event):
    print("Left")
    
def middleClick(event):
    print("Middle")
    
def rightClick(event):
    print("Right")
    
def letterClick(event):
    print("W")
    
frame = tk.Frame(root, width=300, height=250)
# Left click = <Button 1>, Middle Click = <Button 2>, Right Click = <Button 3>
frame.bind("<Button 1>", leftClick)
frame.bind("<Button 2>", middleClick)
frame.bind("<Button 3>", rightClick)
frame.grid(row=0)

root.mainloop()
```

```python
class AlexButtons:
    
    def __init__(self, master):
        frame = tk.Frame(master)
        frame.grid(row=0)
        # common practice to just put everything aligned on the left
        self.printButton = tk.Button(frame, text="Print Message", command=self.printMessage)
        self.printButton.pack(side=tk.LEFT)
        
        self.quitButton = tk.Button(frame, text="Quit", command=frame.quit)
        self.quitButton.pack(side=tk.LEFT)
        
    def printMessage(self):
        print("A message")
        
root = tk.Tk()
alex = AlexButtons(root)
root.mainloop()
```

```python
# Toolbars, drop-down menus, status bars

def doNothing():
    print('Fine.')
    
def makeProject():
    print('Make a project.')
    
def options():
    print('You\'ve found some options.')
    
def startGame():
    print('Starting a game!')

root = tk.Tk()

menu = tk.Menu(root)
root.config(menu=menu)

# Submenu for each sub part
subMenu = tk.Menu(menu)
# Create dropdown menu named File
menu.add_cascade(label="File", menu=subMenu)
# Create option in dropdown menu of File
subMenu.add_command(label="New Project...", command=makeProject)
subMenu.add_command(label="Another Option", command=doNothing)
# Creates a line to separate one group of items from others, visually pleasing
subMenu.add_separator()
subMenu.add_command(label="Exit", command=doNothing)

# Create another menu in our main menu
subMenu2 = tk.Menu(menu)
menu.add_cascade(label="Options", menu=subMenu2)
subMenu2.add_command(label="Redo", command=options)

# Create toolbar with common buttons
toolbar = tk.Frame(root, bg="Orange")

# Remember, first parameter is WHERE DO YOU WANT TO PUT IT
insertButton = tk.Button(toolbar, text="Start New Game", command=startGame)
# Equivalent of float left in web design, can also add padding with "padx", "pady"
insertButton.pack(side=tk.LEFT, padx=2, pady=2)
printButton = tk.Button(toolbar, text="Do nothing", command=doNothing)
printButton.pack(side=tk.LEFT, padx=2, pady=2)
# Apparently they get the same padding

# Now need to display toolbar, doesn't matter that we said top, automatically menu still higher
toolbar.pack(side=tk.TOP, fill=tk.X)

# Now we make a status bar

# Special type of label to appear as a frame; bd = BORDER (can use around labels)
# relief = How you want label / item to appear; SUNKEN is a type of styling for the label
# anchor = Makes sure text is aligned to the WEST or the LEFT
status = tk.Label(root, text="Preparing to do something...", bd=1, relief=tk.SUNKEN, anchor=tk.W)

status.pack(side=tk.BOTTOM, fill=tk.X)

root.mainloop()
```
